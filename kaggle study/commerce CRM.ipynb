CRM 데이터분석은 무엇일까?
    
CRM 데이터분석은 회사의 영업 및 고객 서비스 성과를 보여주는 분석이다. 
CRM분석은 데이터기반의 의사결정을 내리는데 사용할 수 있는 고객 데이터를 제공한다.
기술이 발전하여 CRM분석이 용이했졌다. 예를 들어 CDP등이 그예이다.
    
<span style="color:#31A919;
             font-size:160%;
             font-family:Verdana;">
CRM 분석의 이점
    
CRM 분석의 가장 큰 장점은 영업, 고객 서비스 및 마케팅 프로세스에 정보를 제공하는 데 사용할 수 있다는 점이다. CRM 분석을 사용하여 다음을 통해 방법을 개선할 수 있다. 
다음과 같이 구분된다.:

- **고객 서비스 평가** CRM 분석을 통해 고객 서비스 팀의 성과를 파악할 수 있다. 팀이 개선할 수 있는 수치가 보이면 이러한 목표를 향해 나아갈 수 있는 전략을 구현한다.
- **고객 데이터 도달** 인구통계학적 마케팅에 고객 데이터를 사용하든 이메일 마케팅에 사용하든, 올바른 사람에게 도달하고 있는지 알아야 한다.
- **철저한 고객 분석** 고객이 일반적으로 분기당 얼마를 소비하는가? 고객이 매번 같은 제품을 구매하나요, 아니면 다양하게 구매하는가? CRM 분석을 통해 이러한 질문에 대한 확실한 답을 얻을 수 있으며, 이를 통해 얻은 정보를 마케팅 전략을 개선하는 데 활용할 수 있다.
- **효율적인 리드 생성**: CRM 분석을 통해 어떤 마케팅 활동이 구매와 가장 밀접한 상관관계가 있는지 알 수 있다. 한 가지 접근 방식이 구매와 높은 상관관계를 보이지만 해당 접근 방식으로 일부 고객만 타겟팅했다면 해당 방법을 더 시도해 보자. 매출이 증가할 수 있다.

```python
def replace_with_thresholds(dataframe, variable, q1 = 0.25, q3 = 0.75):
    
    '''
    Detects outliers with IQR method and replaces with thresholds 
    
    '''
    
    df_ = dataframe.copy()
    quartile1 = df_[variable].quantile(q1)
    quartile3 = df_[variable].quantile(q3)
    iqr = quartile3 - quartile1
    
    up_limit = quartile3 + 1.5 * iqr
    low_limit = quartile1 - 1.5 * iqr
    df_.loc[(df_[variable] < low_limit), variable] = low_limit
    df_.loc[(df_[variable] > up_limit), variable] = up_limit
    
    return df_

def ecommerce_preprocess(dataframe):
    df_ = dataframe.copy()
    
    # 결측치
    df_ = df_.dropna()
    
    # 취소 주문, 음수 수량
    df_ = df_[~df_['InvoiceNo'].str.contains('C', na = False)]
    df_ = df_[df_['Quantity'] > 0]
    
    # 이상치 조정
    df_ = replace_with_thresholds(df_, "Quantity", q1 = 0.01, q3 = 0.99)
    df_ = replace_with_thresholds(df_, "UnitPrice", q1 = 0.01, q3 = 0.99)
    
    #Total Price
    df_["TotalPrice"] = df_["Quantity"] * df_["UnitPrice"]
    
    return df_
```

<span style=" font-family:Verdana; font-size:190%; color:#31A919; ">RFM 분석</span>

RFM은 Recency, Frequency, Monetary를 의미한다. 

Recency : 얼마나 최근에 구매했는가
Frequency : 얼마나 자주 구매했는가
Monetary : 얼마나 많은 금액을 지출했는가

즉, 사용자별로 얼마나 최근에, 얼마나 자주, 얼마나 많은 금액을 지출했는지에 따라 사용자들의 분포를 확인하거나 사용자 그룹(또는 등급)을 나누어 분류하는 분석 기법이다.

아래와 같은 사실을 보여준다.

- 최근 구매가 많을수록 프로모션에 대한 고객의 반응이 높다.
- 구매 빈도가 높은 고객일수록 참여도와 만족도가 높다.
- 고액 구매자와 저액 구매자를 차별화한다.

```python
today_date = dt.datetime(2011,12,11)

rfm = df.groupby('CustomerID').agg({'InvoiceDate': lambda x: (today_date - x.max()).days,
                                    'InvoiceNo': lambda x: x.nunique(),
                                    'TotalPrice': lambda x: x.sum()})

rfm.columns = ['recency', 'frequency', 'monetary']
rfm.loc[rfm['monetary'] <= 0, 'monetary'] = 0
rfm = rfm.reset_index()

print(rfm)
```

RFM Scores

```python
def get_rfm_scores(dataframe) -> pd.core.frame.DataFrame:

    df_ = dataframe.copy()
    df_["recency_score"] = pd.qcut(df_["recency"], 5, labels=[5, 4, 3, 2, 1])
    df_["frequency_score"] = pd.qcut(
        df_["frequency"].rank(method="first"), 5, labels=[1, 2, 3, 4, 5]
    )
    df_["monetary_score"] = pd.qcut(df_["monetary"], 5, labels=[1, 2, 3, 4, 5])
    df_["RFM_SCORE"] = df_["recency_score"].astype(str) + df_["frequency_score"].astype(
        str
    )

    return df_


rfm = get_rfm_scores(rfm)
```

```python
def CohortAnalysis(dataframe):

    data = dataframe.copy()
    data = data[["CustomerID", "InvoiceNo", "InvoiceDate"]].drop_duplicates()
    data["order_month"] = data["InvoiceDate"].dt.to_period("M")
    data["cohort"] = (
        data.groupby("CustomerID")["InvoiceDate"].transform("min").dt.to_period("M")
    )
    cohort_data = (
        data.groupby(["cohort", "order_month"])
        .agg(n_customers=("CustomerID", "nunique"))
        .reset_index(drop=False)
    )
    cohort_data["period_number"] = (cohort_data.order_month - cohort_data.cohort).apply(
        attrgetter("n")
    )
    cohort_pivot = cohort_data.pivot_table(
        index="cohort", columns="period_number", values="n_customers"
    )
    cohort_size = cohort_pivot.iloc[:, 0]
    retention_matrix = cohort_pivot.divide(cohort_size, axis=0)
    with sns.axes_style("white"):
        fig, ax = plt.subplots(
            1, 2, figsize=(12, 8), sharey=True, gridspec_kw={"width_ratios": [1, 11]}
        )
        sns.heatmap(
            retention_matrix,
            mask=retention_matrix.isnull(),
            annot=True,
            cbar=False,
            fmt=".0%",
            cmap="coolwarm",
            ax=ax[1],
        )
        ax[1].set_title("Monthly Cohorts: User Retention", fontsize=14)
        ax[1].set(xlabel="# of periods", ylabel="")
        white_cmap = mcolors.ListedColormap(["white"])
        sns.heatmap(
            pd.DataFrame(cohort_size).rename(columns={0: "cohort_size"}),
            annot=True,
            cbar=False,
            fmt="g",
            cmap=white_cmap,
            ax=ax[0],
        )
        fig.tight_layout()
    
CohortAnalysis(df)
```
